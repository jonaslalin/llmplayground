````python
import openai
````

````python
def get_completion(prompt, model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=0.9,
    )
    return response.choices[0].message["content"]
````

````python
joke_subject = "cats"
question_cutoff = 30
answer_cutoff = 20

prompt = f"""
Your task is to come up with a kind joke about {joke_subject}.

To solve the problem, do the following:
1. \
Generate the question word by word. \
Use a scratchpad to keep track of \
each word, \
its length, and \
the total length of the question so far. \
Stop generating new words when the total length of the question exceeds {question_cutoff} characters. \
Ask yourself after each word has been generated if you should continue generating a new word.
2. \
Generate the answer word by word. \
Use a scratchpad to keep track of \
each word, \
its length, and \
the total length of the answer so far. \
Stop generating new words when the total length of the answer exceeds {answer_cutoff} characters. \
Ask yourself after each word has been generated if you should continue generating a new word.
3. \
Output a JSON object with the question and the answer.

Do not output any notes or remarks. \
Only use the following format to solve the problem:
```
SCRATCHPAD FOR THE QUESTION
---
First word in the question: \
<first word in the question>
Length of the first word in the question: \
<length of the first word in the question>
Total length of the question so far: \
<total length of the question so far>
Should I stop generating new words? Is the total length of the question so far (<total length of the question so far>) greater than the stopping criterion of {question_cutoff} characters?: \
<yes or no>
Second word in the question: \
<second word in the question>
Length of the second word in the question: \
<length of the second word in the question>
Total length of the question so far: \
<total length of the question so far>
Should I stop generating new words? Is the total length of the question so far (<total length of the question so far>) greater than the stopping criterion of {question_cutoff} characters?: \
<yes or no>
---

SCRATCHPAD FOR THE ANSWER
---
First word in the answer: \
<first word in the answer>
Length of the first word in the answer: \
<length of the first word in the answer>
Total length of the answer so far: \
<total length of the answer so far>
Should I stop generating new words? Is the total length of the answer so far (<total length of the answer so far>) greater than the stopping criterion of {answer_cutoff} characters?: \
<yes or no>
Second word in the answer: \
<second word in the answer>
Length of the second word in the answer: \
<length of the second word in the answer>
Total length of the answer so far: \
<total length of the answer so far>
Should I stop generating new words? Is the total length of the answer so far (<total length of the answer so far>) greater than the stopping criterion of {answer_cutoff} characters?: \
<yes or no>
---

RESULT
---
{{
    "question": "<the question>",
    "question_total_length": <total length of the question>,
    "answer": "<the answer>",
    "answer_total_length": <total length of the answer>,
    "question_and_answer_total_length": <combined length of the question plus the answer>
}}
---
```

SCRATCHPAD FOR THE QUESTION
"""
response = get_completion(prompt)
print(response)
````
